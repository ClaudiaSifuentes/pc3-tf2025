```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
try:
    df = pd.read_csv('promedios.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('promedios.csv', encoding='latin-1')

df.head(10)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
temp_df1 = df.drop(columns=['TIPO_DISCAPACIDAD'])
temp_df2 = temp_df1.dropna(how='any', axis=0)
temp_df2.head(10)
```

```python
def convert_year_semester_column(df, column_name):
    df = df.copy()
    df[column_name] = df[column_name].astype(str).str.replace('-', '', regex=False)
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    return df

temp_df3 = convert_year_semester_column(temp_df2, 'PERIODO_ACADEMICO')
temp_df4 = temp_df3.drop(columns=['Periodo de Ingreso', 'MODALIDAD_ESTUDIO', 'PROMEDIO_PONDERADO_ACUMULADO_AL_PERIODO_ANTERIOR', 'PERIODO_ACADEMICO'])
temp_df4.head(10)
```

```python
temp_df5 = temp_df4.copy()
temp_df5['FECHA_MATRICULA'] = pd.to_datetime(temp_df4['FECHA_MATRICULA'], format='%d/%m/%Y')
temp_df5['GENERO'] = temp_df5['GENERO'].map({'FEMENINO': True, 'MASCULINO': False})
temp_df5['DISCAPACIDAD'] = temp_df5['DISCAPACIDAD'].map({'SI': True, 'NO': False})
```

```python
temp_df6 = temp_df5.copy()

# Convert dates with error handling for invalid/out-of-bounds dates
temp_df6['Fecha Nacimiento'] = pd.to_datetime(temp_df6['Fecha Nacimiento'], 
                                            format='%d/%m/%Y', 
                                            errors='coerce')  # Convert invalid dates to NaT

# Check for any invalid dates that were converted to NaT
invalid_dates = temp_df6['Fecha Nacimiento'].isna()
temp_df6 = temp_df6[~invalid_dates]  # Remove rows with invalid dates
if invalid_dates.any():
    print(f"Found {invalid_dates.sum()} invalid birth dates that were converted to NaT")
    
# Calculate age only for valid dates
temp_df6['Edad'] = (temp_df6['FECHA_MATRICULA'] - temp_df6['Fecha Nacimiento']).dt.days / 365.25
temp_df6 = temp_df6[temp_df6['Edad'] >= 14]  # Filter out ages less than 14
temp_df6.drop(columns=['Fecha Nacimiento'], inplace=True)  # Drop the birth date column
# Display some statistics about the ages
print(f"Age statistics (in years):")
print(temp_df6['Edad'].describe())
```

```python
temp_df7 = pd.get_dummies(temp_df6, columns=['Programa'], prefix='Programa')
temp_df7 = pd.get_dummies(temp_df7, columns=['Facultad'], prefix='Facultad')
```
